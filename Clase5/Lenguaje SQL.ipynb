Se puede realizar los diferentes tipos de pegados, por ejemplo un INNER JOIN:

```python
q5="""SELECT s.ESTU_CONSECUTIVO, s.COLE_COD_MCPIO_UBICACION, s.PUNT_INGLES,
             d.cod_mpio, d.nom_mpio
FROM s11_2019I s 
JOIN divipola d
ON s.COLE_COD_MCPIO_UBICACION = d.cod_mpio;"""
icfes_2019I=sqldf(q5)
icfes_2019I
```

También se pueden realizar filtros: para las personas de bogotá que realizaron las pruebas saber en el primer semestre del 2019 calcular el promedio de matemáticas POR JORNADA Y NATURALEZA:

```python
q6="""SELECT COLE_JORNADA, COLE_NATURALEZA, AVG(PUNT_MATEMATICAS) AS 'PROM_MAT',
COUNT(*) AS N
FROM s11_2019I
WHERE COLE_COD_MCPIO_UBICACION = '11001'
GROUP BY COLE_JORNADA, COLE_NATURALEZA
ORDER BY PROM_MAT DESC
"""
consulta=sqldf(q6)
consulta
```
